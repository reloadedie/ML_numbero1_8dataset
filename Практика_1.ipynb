```python
# Столбчатые диаграммы для категориальных переменных
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
axes = axes.flatten()

# 1. Распределение по полу с разделением по расстройствам сна
sns.countplot(data=df, x='Gender', hue='Sleep Disorder', ax=axes[0])
axes[0].set_title('Распределение по полу с разделением по расстройствам сна')
axes[0].set_ylabel('Количество')

# 2. Распределение по категориям BMI с разделением по расстройствам сна
sns.countplot(data=df, x='BMI Category', hue='Sleep Disorder', ax=axes[1])
axes[1].set_title('Распределение по BMI с разделением по расстройствам сна')
axes[1].set_ylabel('Количество')
axes[1].tick_params(axis='x', rotation=45)

# 3. Распределение по профессиям (топ-6 самых частых)
top_occupations = df['Occupation'].value_counts().head(6).index
df_top_occ = df[df['Occupation'].isin(top_occupations)]
sns.countplot(data=df_top_occ, x='Occupation', hue='Sleep Disorder', ax=axes[2])
axes[2].set_title('Топ-6 профессий с разделением по расстройствам сна')
axes[2].set_ylabel('Количество')
axes[2].tick_params(axis='x', rotation=45)

# 4. Распределение по качеству сна с разделением по расстройствам
sns.countplot(data=df, x='Quality of Sleep', hue='Sleep Disorder', ax=axes[3])
axes[3].set_title('Качество сна с разделением по расстройствам')
axes[3].set_ylabel('Количество')

plt.tight_layout()
plt.show()
```

Концентрации переменных:

```python
# Более детальный анализ по профессиям и расстройствам сна
plt.figure(figsize=(12, 6))
occupation_disorder = pd.crosstab(df['Occupation'], df['Sleep Disorder'])
occupation_disorder.plot(kind='bar', stacked=True)
plt.title('Распределение расстройств сна по профессиям')
plt.ylabel('Количество людей')
plt.xlabel('Профессия')
plt.xticks(rotation=45)
plt.legend(title='Расстройство сна')
plt.tight_layout()
plt.show()

# Процентное распределение
print("ПРОЦЕНТНОЕ РАСПРЕДЕЛЕНИЕ РАССТРОЙСТВ СНА ПО ПРОФЕССИЯМ:")
print("=" * 50)
occupation_percent = (occupation_disorder.div(occupation_disorder.sum(axis=1), axis=0) * 100).round(1)
print(occupation_percent)
```

**Боксплот для числовой переменной:**

```python
# Боксплоты с разделением по Sleep Disorder
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
axes = axes.flatten()

# Сравнение по наличию расстройств сна
sns.boxplot(data=df, x='Sleep Disorder', y='Sleep Duration', ax=axes[0])
axes[0].set_title('Длительность сна по типам расстройств')

sns.boxplot(data=df, x='Sleep Disorder', y='Quality of Sleep', ax=axes[1])
axes[1].set_title('Качество сна по типам расстройств')

sns.boxplot(data=df, x='Sleep Disorder', y='Stress Level', ax=axes[2])
axes[2].set_title('Уровень стресса по типам расстройств')

sns.boxplot(data=df, x='Sleep Disorder', y='Physical Activity Level', ax=axes[3])
axes[3].set_title('Физ. активность по типам расстройств')

plt.tight_layout()
plt.show()
```

**Постройте корреляционную матрицу для всех числовых переменных и визуализируйте её с помощью тепловой карты:**

```python
# Сначала удалим нечисловые колонки для корреляции
df_numeric = df.select_dtypes(include=['int64', 'float64'])

# Корреляционная матрица
correlation_matrix = df_numeric.corr()

# Визуализация тепловой карты
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0,
           square=True, fmt='.2f')
plt.title('Корреляционная матрица числовых переменных')
plt.tight_layout()
plt.show()
```
